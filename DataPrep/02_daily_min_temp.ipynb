```python
data['Date_']=pd.to_datetime(data['Date'],format="%d/%m/%Y")
```

```python
data.info()
```

```python
#date time features 
```

```python
data['month']=data['Date_'].apply(lambda x:x.month)
data['day']=data['Date_'].apply(lambda x:x.day)

data.head(),data.shape
```

```python
#lag features
```

```python
df=pd.concat([data['temp'].shift(3),data['temp'].shift(2),data['temp'].shift(1),data['temp']],axis=1)
df.columns = ['t-2', 't-1', 't', 't+1']
df.head()
```

```python
# rolling window statistics
```

```python
data=data[:100]
```

```python
data
```

```python
temps=data[['temp']].astype(str).astype(float)
width = 3
shifted = temps.shift(width-1) # use 1 for actual rolling window 
window = shifted.rolling(window=width)
dataframe = pd.concat([window.min(), window.max(),temps],axis=1)
dataframe.columns = ['min', 'max', 't+1']
dataframe[:20]
```

```python
# expanding window
```

```python
window = temps.expanding()
dataframe = pd.concat([window.min(), window.mean(), window.max(), temps.shift(-1)], axis=1)
dataframe.columns = ['min', 'mean', 'max', 't+1']
```
